```python
obo_metadata = "http://obofoundry.org/registry/ontologies.yml"
categories_yml = "obocategories.yml"
```

```python
import yaml

with open(categories_yml, 'r') as stream:
    try:
        cat=yaml.safe_load(stream)
    except yaml.YAMLError as exc:
        print(exc)
```

```python
import urllib.request
data = urllib.request.urlopen(obo_metadata)
try:
    obo=yaml.safe_load(data)
except yaml.YAMLError as exc:
    print(exc)
```

```python
def get_category(old_domain,categories):
    for category in categories['categories']:
        if category['category']==old_domain:
            return category['category']
        if 'subsume' in category:
            for tag in category['subsume']:
                if tag==old_domain:
                    return category['category']
    return "other"
```

```python
data = []
for o in obo['ontologies']:
    if 'domain' in o:
        #print("%s (%s): %s" % (o['domain'],o['id'],get_category(o['domain'],cat)))
        row = [o['domain'],o['id'],get_category(o['domain'],cat)]
        data.append(row)
        
import pandas as pd

df = pd.DataFrame(data, columns=['domain','ontology','category'])
df = df.sort_values(by=['category','ontology'])
df[['category','ontology','domain']].to_csv("obo_new_categories.tsv",sep="\t",index=False)
```

```python
df_summary=df[['category','ontology']].groupby('category').count()
df_summary=df_summary.add_suffix('_count').reset_index()
df_summary=df_summary.sort_values(by=['ontology_count'],ascending=False)
df_summary.reset_index(inplace=True)
```
